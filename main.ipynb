```python
import gc

import pandas as pd
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone_Design2/dataset/big_data_set1_f.csv', encoding='cp949')
data2 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone_Design2/dataset/big_data_set2_f.csv', encoding='cp949')
data3 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone_Design2/dataset/big_data_set3_f.csv', encoding='cp949')

data = pd.merge(data2, data3, on=['ENCODED_MCT' ,'TA_YM'], how='inner')
data = pd.merge(data, data1, on='ENCODED_MCT', how='left')

del data1, data2, data3
gc.collect()

data
```

```python
data_prep = data.drop(columns=['MCT_SIGUNGU_NM','MCT_BSE_AR'])
data_prep
```

```python
data_prep.info()
```

```python
ob_lst = []
flt_lst = []
int_lst = []
for c in data_prep.columns:
    if data_prep[c].dtype == 'O':
        ob_lst.append(c)
    elif data_prep[c].dtype == 'float':
        flt_lst.append(c)
    else:
        int_lst.append(c)
print(f'int_col num : {len(int_lst)}', f'float_col num : {len(flt_lst)}', f'object_col num : {len(ob_lst)}',
      sep='\n')
```

```python
cols = ['MCT_OPE_MS_CN','RC_M1_SAA','RC_M1_TO_UE_CT','RC_M1_UE_CUS_CN','RC_M1_AV_NP_AT','APV_CE_RAT']
for cc in cols:
    for idx in data_prep.index:
        data_prep.loc[idx, cc] = data_prep.loc[idx, cc].split('_')[0]
data_prep
```
