## Data Preprocessing

```python
import gc

import pandas as pd
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone2_HelloKitty/dataset/big_data_set1_f.csv', encoding='cp949')
data2 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone2_HelloKitty/dataset/big_data_set2_f.csv', encoding='cp949')
data3 = pd.read_csv('/content/drive/MyDrive/B.S.degree/Capstone2_HelloKitty/dataset/big_data_set3_f.csv', encoding='cp949')

data = pd.merge(data2, data3, on=['ENCODED_MCT' ,'TA_YM'], how='inner')
data = pd.merge(data, data1, on='ENCODED_MCT', how='left')

del data1, data2, data3
gc.collect()

data
```

```python
data['MCT_BSE_AR'].value_counts()  ## drop 확정
```

```python
data_prep = data.drop(columns=['MCT_BRD_NUM', 'HPSN_MCT_BZN_CD_NM', 'MCT_SIGUNGU_NM', 'MCT_NM', 'MCT_BSE_AR', 'MCT_BSE_AR'])
data_prep
```

```python
data_prep.info()
```

```python
columns_to_split = ['MCT_OPE_MS_CN', 'RC_M1_SAA', 'RC_M1_TO_UE_CT','RC_M1_UE_CUS_CN','RC_M1_AV_NP_AT','APV_CE_RAT']

for c in columns_to_split:
    mask = data_prep[c].notna()
    data_prep.loc[mask, c] = data_prep.loc[mask, c].astype(str).str.split('_').str[0].astype(int)
data_prep
```
